```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Polynomial regression
* can fit non-linear dataset
* can find relations between features
* if there's $a, b$ features & we use `degree=3` it will make new set of features $a, a^2, a^3, ab, a^2b, b, b^2, b^3, b^2a$
* if there's $n$ features and need $d$ dimensions it will generate $\frac{(n+d)!}{n!d!}$ so take care from ___factorial explosion___

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
plt.plot(X, y, 'b.')
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
print(X[0], X_poly[0]) #[x_0], [x_0, x_0^2]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression().fit(X_poly, y)
#0.5 * X**2 + X + 2
lin_reg.coef_, lin_reg.intercept_ # [1, 0.5], 2
```

```python
plt.plot(X, lin_reg.predict(X_poly), 'r+')
plt.plot(X, y, 'b.')
```
